```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
```

```python
df= df.drop_duplicates()
```

```python
df = df.dropna()
```

```python
sns.boxplot(x=df["Price"])
plt.title("Price")
plt.show()
```

```python
sns.boxplot(df["HP"])
plt.title("Horse Power")
plt.show()
```

```python
df.Make.value_counts().nlargest(40).plot(kind="bar",figsize=(10,5))
plt.title("Number of cars by DERR")
plt.ylabel("Number of Cars ")
plt.xlabel("Make")
plt.show()
```

```python
df.Year.value_counts().nlargest(15).plot(kind="bar",figsize=(10,5),legend=True)
plt.title("Yearly sales")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend(["Transmission"])
plt.show()
```

```python
group=df.groupby("Model")["Make"].count().sort_values(ascending=False).head(20)
group.plot(kind="bar",figsize=(12,6),width=0.3)
plt.title("Maximum Price For Each Model")
plt.xlabel("Model Name")
plt.ylabel("Price ")
plt.show()
```

```python
group=df.groupby("Drive Mode")["Make"].count()
explode = (0.2, 0.2, 0.2, 0.2)
group.plot(kind="pie",figsize=(10,6),explode = explode,autopct='%1.1f%%')
plt.title(" Drive Mode and count of cars")
plt.ylabel("")
plt.show()
```
